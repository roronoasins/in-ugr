```python
import pandas as pd
subidas = pd.read_csv("kaggle-subs.csv", na_values='-')
```

```python
subidas
```

```python
subidas.columns = [subidas['Unnamed: 0'][0], subidas['Unnamed: 1'][0], subidas['Unnamed: 2'][0], subidas['Unnamed: 3'][0], subidas['Unnamed: 4'][0], subidas['Unnamed: 5'][0]]
subidas = subidas.drop(0)
subidas['Private Score'] = subidas['Private Score'].str.replace(',','.')
subidas['Local AUC'] = subidas['Local AUC'].str.replace(',','.')
subidas['Local Acc'] = subidas['Local Acc'].str.replace(',','.')
```

```python
subidas
```

```python
private = subidas.loc[:,'Private Score'].values.astype('float')
local_auc = subidas.loc[:,'Local AUC'].values.astype('float')
local_acc = subidas.loc[:,'Local Acc'].values.astype('float')
number = subidas.loc[:,'Submission'].values.astype('float')
```

```python
from matplotlib import pyplot as plt
plt.plot(number, private, label='kaggle score')
plt.plot(number, local_auc, label='local auc')
plt.plot(number, local_acc, label='local acc')
plt.xlabel("N submission")
plt.ylabel("Score")
plt.legend()
plt.xticks(number)
plt.show()
```
